# Initialize "local" git repo on google drive

Only need to run this once per repo at the begining when you start developing using this workflow

```
from google.colab import drive
drive.mount('/content/gdrive')

# cd into the folder where you want the git repo to be located
%cd gdrive/MyDrive/path-to-parent-folder-on-gdrive

!git config --global user.email 'You@Your.com'
!git config --global user.name 'Username'

#clone repo from github
!git clone https://github.com/<username>/<reponame>.git
```

```
%cd /content/gdrive/MyDrive/path-to-repo
!git pull
!ls
```

# Git commands for colab workflow

Connect google drive, enter repo and pull most recent repo

```
from google.colab import drive
drive.mount('/content/gdrive')

%cd /content/gdrive/MyDrive/path-to-repo

!git config --global user.email 'You@Your.com'
!git config --global user.name 'Username'

!git pull
```

Add all the files like

```
!git add -A
```

then commit

```
!git commit -m 'my first commit with colab'
```

this is the most important line, when you add the git remote origin you need to pass your Github credentials with the URL like,

```
!git remote add origin https://<username>:<password>@github.com/<username>/reponame.git
```

if you got “fatal: remote origin already exists” error try running this command

```
!git remote rm origin
```

at last, push your awesomeness

```
!git push -u origin <branch>
```
