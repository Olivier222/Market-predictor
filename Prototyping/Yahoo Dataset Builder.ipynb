```
from yahooquery import Ticker
import json
import pandas as pd

# get list of S&P500 symbols
table=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df = table[0]
symbols = list(df['Symbol'])
sp500 = Ticker(symbols, asynchronous=True)

all_data = sp500.all_modules

for d in all_data['FB']['cashflowStatementHistoryQuarterly']['cashflowStatements']:
  print(d['endDate'])

print(all_data['FB']['summaryDetail'])
print(all_data['FB']['incomeStatementHistory']['incomeStatementHistory'])

json_file = os.path.join(base_dir, 'SP500_Yahoo_IndividualData.json')
with open(json_file, 'w') as fp:
    json.dump(all_data, fp, indent=4)

with open(json_file, 'r') as fp:
    all_modules = json.load(fp)

for d in all_modules['FB']['cashflowStatementHistoryQuarterly']['cashflowStatements']:
  print(d['endDate'])

print(all_modules['FB']['summaryDetail'])
print(all_modules['FB']['incomeStatementHistory']['incomeStatementHistory'])


set(all_data) ^ set(all_modules)
```
